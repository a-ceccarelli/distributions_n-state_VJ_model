```python
"""This code takes some time to run due to the high number of sampled data points
Decrease T = 100000 for quicker but less accurate empirical probabilities"""

#importing useful libraries
import numpy as np
import matplotlib.pyplot as plt

#getting current directory
import os
cwd = os.getcwd()

#importing all the functions contained in "functions.py"
%run ../functions
```

```python
#time lengths
delta_t = 0.3 
T = 100000

#define parameters of the two-state model
V = np.array([2000, -1500])
Lambda = np.array([1, 0.5])
P = np.array([[0.0, 1], [1, 0.0]])
sigma = 50
n = Lambda.shape[0]

#plotting the figure
plt.figure(figsize=(3, 1.5))

for i in [1, 5, 10]:
    #running the model
    x_sim, y_sim, t_sim, sw_true, st = n_state_model(delta_t, T, V, Lambda*i, P, sigma)
    
    Tot_dt = sw_true.shape[0]
    n_sw_max = int(np.max(sw_true))
    sw_count = np.zeros(n_sw_max+1)
    for sw_num in sw_true:
        sw_count[int(sw_num)] += 1
    sw_probs = sw_count/Tot_dt

    plt.plot(np.arange(n_sw_max+1), sw_probs, '-*',
                 label=str(r'$i=$ '+str(i)))

plt.legend()
plt.xlabel(r'$w$')
plt.ylabel('$P(W=w)$')
plt.ylim([-0.01, 1.01])
plt.xticks(np.arange(0, n_sw_max+1))
plt.savefig("P_k_switches_lambdas_2.png", format="png", dpi=1200,
            bbox_inches="tight") 
plt.show()
```
